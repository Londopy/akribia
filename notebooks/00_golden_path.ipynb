# akribia — the thesis in 90 seconds

**One inference engine. Three miscalibrations. Same math.**

Predictive coding treats the brain as a hierarchical inference machine: a prediction meets sensory evidence, the mismatch is a *prediction error*, and that error is weighted by *precision* (how much the system trusts the signal) before updating beliefs. The same precision-weighting math, miscalibrated at different points, produces autism, ADHD, and PPCS. Below: the *same* engine and the *same* task, run under two precision parameterizations, produce different perceptual behaviour.

> Not a diagnostic tool. See `docs/LIMITATIONS.md`.

```python
from akribia import core, profiles
from akribia.tasks import illusion_task
print('backend:', core.BACKEND)
```

## Baseline: a Kanizsa illusory contour

The neurotypical reference 'sees' the illusory triangle: a positive illusion-susceptibility score.

```python
baseline = illusion_task.run(profiles.get('baseline'))
baseline['summary']['illusion_score']
```

## Same engine, weak-prior parameterization

`autism_weak_prior` caps prior precision, so absent local evidence at the illusory midpoints dominates — the literature's *reduced illusion* finding.

```python
weak = illusion_task.run(profiles.get('autism_weak_prior'))
weak['summary']['illusion_score']
```

## The single image that demonstrates the whole project

```python
import matplotlib
matplotlib.use('Agg')
fig = illusion_task.plot(weak)
fig
```

That's the whole idea — **same engine, different precision parameterization, different perceptual behaviour**. The other profiles miscalibrate precision at other points (reward valuation → ADHD, forward model → PPCS), and `comorbid` / `audhd_emotion_dysregulation` run more than one at once.
