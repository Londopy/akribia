# Autism — the contested precision debate

Two competing accounts, both shipped as switchable profiles (see `docs/THEORY.md` §2). `autism_weak_prior` (hypo-priors) vs. `autism_overfitting` (HIPPEA / inflexible precision). The illusion task discriminates the weak-prior account; the volatility task discriminates HIPPEA.

## Illusion susceptibility — weak priors reduce it

```python
from akribia import profiles
from akribia.tasks import illusion_task
for p in ['baseline','autism_weak_prior','autism_overfitting']:
    s = illusion_task.run(profiles.get(p))['summary']
    print(f"{p:18s} illusion_score={s['illusion_score']:.3f}")
```

## Volatility switch — HIPPEA reconverges more slowly (a *transient* delay)

```python
from akribia.tasks import volatility_learning_task as v
for p in ['baseline','autism_overfitting']:
    s = v.run(profiles.get(p))['summary']
    print(f"{p:18s} trials_to_reconverge={s['trials_to_reconverge']:.2f}")
```
